```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from matplotlib import rc
rc('font', family='Malgun Gothic') # 한글 폰트
plt.rcParams['axes.unicode_minus'] = False # 마이너스 부호출력 설정
plt.rc('font', size=16)

import warnings
warnings.filterwarnings('ignore')
```

```python
import seaborn as sns
```

```python
df = pd.read_pickle('./건강검진그룹.pkl')
```

```python
vals = df.gluCAT_1n2_r.values
```

```python
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111)

ax.plot(range(len(vals)), vals, color='gray', alpha=0.3)
ax.scatter(range(5), vals[:5], color='purple')
ax.scatter(range(5, 8), vals[5:8], color='blue')
ax.scatter(range(8, 13), vals[8:13], color='green')
ax.scatter(range(13, 18), vals[13:18], color='orange')
ax.scatter([18], vals[-1], color='red')

ax.set_ylabel('당뇨전단계+당뇨 비율(%)')
ax.set_xlabel('소그룹 번호')
ax.set_xticks(range(0, 20))
ax.grid(axis='y')

plt.show()
```
