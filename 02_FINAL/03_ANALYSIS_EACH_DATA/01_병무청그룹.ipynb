* 전단계를 정상과 바꿔적음을 확인하여 수정한 부분

```python
df['진짜 전단계'] = df['전체 카운트'] - df['전단계 카운트'] - df['당뇨 카운트']
```

```python
df.drop(columns=['전단계 카운트'], inplace=True)
```

```python
df
```

```python
df['glu1_r'] = (df['진짜 전단계']/df['전체 카운트']*100).round(2)
```

```python
df['glu2_r'] = (df['당뇨 카운트']/df['전체 카운트']*100).round(2)
```

```python
df['glu_totr'] = df.glu1_r+df.glu2_r
```
